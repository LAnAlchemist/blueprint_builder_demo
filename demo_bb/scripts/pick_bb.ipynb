#### Filter based the score terms and/or the filters
- Filter the designed proteins based on the previoius scores. Usually these score cut-offs are scaffold-dependent, it is suggested to take a reseanable percetage cutoff based on previous distributions for each score term, instead of focusing on a specific number of the scores.
- The information of the designed backbones passing the filter is printed at the end of this cell.

```python
VDW_CUTOFF = 10.0
RAMA_CUTOFF = 10.0
OMEGA_CUTOFF = 20.0
HBOND_LR_BB = -60.0
HBOND_SR_BB = -20.0
df_sel = merged_df.loc[(merged_df['vdw'] < VDW_CUTOFF) &
                       (merged_df['cen_rama'] < RAMA_CUTOFF) &
                       (merged_df['omega'] < OMEGA_CUTOFF) &
                       (merged_df['hbond_lr_bb'] < HBOND_LR_BB) &
                       (merged_df['hbond_sr_bb'] < HBOND_SR_BB)]
df_sel #this line displaies the scores.
```

#### Another more straightforward way to display all score terms is to make a histogram figure for each score term
- Only the selected core score terms are explained in this tutorial, the explanation of other score terms can be found at https://www.rosettacommons.org/docs/wiki/rosetta_basics/scoring/score-types
- Some useful but not introduced score terms are listed in the relevant_features.
- In general, we don't have a hard cutoff recommendation for most score terms, it is recommended to check the distribution of the score terms and take the top fraction as cutoff.
- Please note at this point, the protein backbones are actually still polyalanine peptides, some score terms may give significiantly unideal scores.

```python
relevant_features = ['total_score', 'angle_constraint', 'atom_pair_constraint',\
       'cen_rama', 'fa_atr', 'fa_dun', 'fa_elec','fa_intra_rep', 'fa_rep', 'fa_sol',\
       'hbond_bb_sc', 'hbond_lr_bb', 'hbond_sc', 'hbond_sr_bb','omega', 'p_aa_pp', 'pro_close',\
       'rama_prepro', 'rg', 'rsigma', 'sspair','vdw']
ncols = 4
nrows = math.ceil(len(relevant_features) / ncols)
(fig, axs) = plt.subplots(
    ncols=ncols, nrows=nrows, figsize=[18,3*nrows]
)
axs = axs.reshape(-1)

for (i, metric) in enumerate(relevant_features):
    sns.distplot(df_sel[metric], ax=axs[i])
    
sns.despine()
plt.tight_layout()
plt.show()
```
