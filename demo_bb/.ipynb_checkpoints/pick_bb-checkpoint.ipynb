#### Read the score file

```python
#get score list
df_s = []
for score_fn in score_s:
    df = pd.read_csv('%s'%score_fn, sep='\s+', header=1)
    df_s.append(df)
#
merged_df = pd.concat(df_s)
```

#### Filter the designed proteins based on the previoius scores. Usually these score cut-offs are scaffold-dependent, it is suggested to take a reseanable percetage cutoff based on previous distributions for each score term, instead of focusing on a specific number of the scores.
#### The information of the designed backbones passing the filter is printed at the end of this cell.

```python
VDW_CUTOFF = 10.0
RAMA_CUTOFF = 10.0
OMEGA_CUTOFF = 20.0
HBOND_LR_BB = -60.0
HBOND_SR_BB = -25.0
df_sel = merged_df.loc[(merged_df['vdw'] < VDW_CUTOFF) &
                       (merged_df['cen_rama'] < RAMA_CUTOFF) &
                       (merged_df['omega'] < OMEGA_CUTOFF) &
                       (merged_df['hbond_lr_bb'] < HBOND_LR_BB) &
                       (merged_df['hbond_sr_bb'] < HBOND_SR_BB)]
df_sel
```

#### Create a new folder to contain all backbones passing the filter for further sequence design

```python
os.makedirs(WORK_DIR+'/sel_for_seq_design',exist_ok=True)
```

#### Copy the selected backbones to the new folder

```python
for pdb in df_sel['description']:
    pdb_path = PDB_DIR+'/'+pdb+'.pdb'
    cmd = 'cp -n {} {}'.format(pdb_path,WORK_DIR+'/sel_for_seq_design')
    subprocess.call('{}'.format(cmd),shell=True)
```
